## Import Libraries

```python
import pandas as pd
from sklearn.tree import plot_tree,DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

## Reading the dataset

```python
dataset=pd.read_csv("iris.csv",index_col=0)
```

```python
dataset.head()
```

## Extracting the features from dataset

```python
X=dataset.iloc[:,0:4]
Y=dataset.iloc[:,4]
```

```python
X
```

## Splitting The Dataset into Training(70%) and Testing(*30*%) data

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,train_size=0.70,random_state=2)
```

```python
decision_tree=DecisionTreeClassifier()
```

## Training the model

```python
decision_tree.fit(xtrain,ytrain)
```

## The accuracy of the model

```python
print("The accuracy of model is ",decision_tree.score(xtest,ytest))
```

## visualization of tree

```python
n=[]
n=dataset.columns
```

```python
fig=plt.figure(figsize=(25,20))
plot_tree(decision_tree,feature_names=n[[0,1,2,3]],class_names=n[4],filled=True)
```
